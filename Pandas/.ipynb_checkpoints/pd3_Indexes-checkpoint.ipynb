## Setting an Indexes, generally for columns having unique values like email

```python
df.set_index('Hobbyist')
```

#### But if we get df again, those index no longer exists

```python
df
```

#### To set index permanently, we can do as 

```python
df.set_index('Hobbyist', inplace=True)
```

```python
df
```

```python
df.index
```

### This setting of index is useful to get information from that perticular index value, earlier in loc method we are passing with index number but now we can pass actual value

```python
df.loc['Yes']
```

### But iloc method is still available to get data using index numbers, with loc it is not possible

```python
df.iloc[0]
```

### To reset index 

```python
df.reset_index(inplace=True)
```

```python
df
```

### Deafult index can be also passed while reading the csv, using index_col

```python
schema = pd.read_csv('survey_results_schema.csv', index_col='Column')
```

```python
schema
```

```python
schema.loc['Hobbyist']
```

```python
schema.loc['Hobbyist','QuestionText']
```

### To sort index

```python
schema.sort_index()
```

```python
### To reverse the order of index
```
