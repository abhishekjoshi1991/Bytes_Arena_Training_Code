```python
import pandas as pd
```

```python
people = {'name' : ['abhi','john', 'mark', 'harry'],
         'surname': ['joshi','doe', 'doe', 'potter'],
         'Hobby':['play','sing','dance','magic']}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# Sorting Data
# use sort values method, pass name of columns to sort
# multiple columns can be passed while sorting
# order (ascending true or false) can be specified
```

### To sort results

```python
df.sort_values(by='surname')
# default sorting by ascending
```

```python
# to sort in descending order, set ascending to false
```

```python
df.sort_values(by='surname', ascending=False)
```

```python
# To sort data by surname first and then name, that is sorting on multiple columns
```

```python
df.sort_values(by=['surname','name'])
```

```python
# Suppose we want to sort surname column in descending order and name column in ascending order, then
# we can pass list of true false values that corresponds to column name
```

```python
df.sort_values(by=['surname','name'], ascending=[False, True])
```

```python
# so when it finds to same surname i.e. 'doe' then it sorts name in ascending order
```

```python
# To make changes permanent, use inplace=True
```

```python
df.sort_values(by=['surname','name'], ascending=[False, True], inplace=True)
```

```python
df
```

### Sort Index method

```python
# to reset index back to normal , from 0 to ...., rest_index method is used or to sort with index
```

```python
df.sort_index()
```

```python
df.sort_index(ascending=False)
```
